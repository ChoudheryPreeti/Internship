```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
```

```python
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
driver.maximize_window()
```

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
Rank = []
Name = []
Artist = []
Date = []
Views = []
```

```python
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]"):
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append("-")
        
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]"):
        Name.append(i.text)
except NoSuchElementException:
    Name.append("-")
        
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]"):
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")       

try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]"):
        Date.append(i.text)
except NoSuchElementException:
    Date.append("-")
        
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]"):
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")
```

```python
print(len(Rank),len(Name),len(Date),len(Artist),len(Views))
```

```python
df=pd.DataFrame({'Rank':Rank,'Name':Name,'Date':Date,'Artist':Artist,'Views':Views})
df
```

```python
Trending=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a")
Trending.click()
```

```python
URL=[]
url=driver.find_elements(By.XPATH,"//a[@class='Link']")
for i in url:
    URL.append(i.get_attribute("href")) 
```

```python
Repository_title = []
Repository_description  = []
Contributors_count  = []
Language_used = []
LG=[]
```

```python
title = driver.find_elements(By.XPATH,"//a[@class ='Link']")
for i in title:
    Repository_title.append(i.text)
Repository_title   

for i in URL:
    driver.get(i)
   
    try:
        description= driver.find_element(By.XPATH,"//p[@class='f4 my-3']")
        Repository_description.append(description.text)
    except NoSuchElementException:
        Repository_description.append('-')
    
    try:
        contributor=driver.find_element(By.XPATH,"//*[contains(text(),'Contributors')]")
        Contributors_count.append(contributor.text.replace('Contributors','').replace('\n',''))
    except NoSuchElementException:
        Contributors_count.append('-')
    
    try:
        for i in driver.find_elements(By.XPATH,"//span[@class='color-fg-default text-bold mr-1']"):
            LG.append(i.text.replace('\n',' '))
        Language_used.append(LG)
    except NoSuchElementException:
        Language_used.append('-')
```

```python
try:
    names=driver.find_elements(By.XPATH,'//div[@class="o-chart-results-list-row-container"]//h3')[::4]
    for i in names:
        Song_name.append(i.text)
except NoSuchElementException:       
    Song_name.append('-')
    
try:
    artist=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]//span')[::7]
    for i in artist:
        Artist_name.append(i.text)
except NoSuchElementException:     
    Artist_name.append('-')
    
try:
    last_week=driver.find_elements(By.XPATH,'//li[@class="o-chart-results-list__item // a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-background-color-white-064@mobile-max u-hidden@mobile-max"]')[::2]
    for i in last_week:
        Last_week_rank.append(i.text)
except NoSuchElementException:      
    Last_week_rank.append('-')
    
try:
    peak=driver.find_elements(By.XPATH,'//li[@class="o-chart-results-list__item // a-chart-bg-color a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-background-color-grey-lightest lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-hidden@mobile-max"]')[1::2]
    for i in peak:
        Peak_rank.append(i.text)
except NoSuchElementException:       
    Peak_rank.append('-')
    
try:
    weeks=driver.find_elements(By.XPATH,'//li[@class="o-chart-results-list__item // a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-background-color-white-064@mobile-max u-hidden@mobile-max"]')[1::2]
    for i in weeks:
        Weeks_on_board.append(i.text)
except NoSuchElementException:      
    Weeks_on_board.append('-') 
```

```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
```

```python
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
driver.maximize_window()
```

```python
url = "https://archive.ics.uci.edu/"
driver.get(url)
```

```python
view_dataset=driver.find_element(By.XPATH,"//a[@class='btn-primary btn']")
view_dataset.click()
```

```python
URL=[]
url=driver.find_elements(By.XPATH,"//a[@class='link-hover link text-xl font-semibold']")
for i in url:
    URL.append(i.get_attribute("href"))
```

```python
print(len(URL))
```

```python
Dataset_Name=[]
Data_Type=[]
Task=[]
Attribute_Type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
for i in URL:
    driver.get(i)
    
    try:
        dataset_name = driver.find_element(By.XPATH,"//h1[@class='text-3xl font-semibold text-primary-content']")
        Dataset_Name.append(dataset_name.text)
    except NoSuchElementException:
        Dataset_Name.append('-')

    try:
        Data_type = driver.find_element(By.XPATH,"//p[@class='text-md']")
        Data_Type.append(Data_type.text)
    except NoSuchElementException:
        Data_Type.append('-')

    try:
        task= driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p")
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('-')
    
    try:
        attribute_type=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p")
        Attribute_Type.append(attribute_type.text)
    except NoSuchElementException:
        Attribute_Type.append("-")
    
    try:
        no_of_instances= driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p")
        No_of_instances.append(no_of_instances.text)
    except NoSuchElementException:
         No_of_instances.append('-')
    
    try:
        no_of_attribute= driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p")
        No_of_attribute.append(no_of_attribute.text)
    except NoSuchElementException:
        No_of_attribute.append('-')
    
    try:
        year= driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2")
        Year.append(year.text.split(' ')[2])
    except NoSuchElementException:
        Year.append('-')
        
```

```python
print(len(Dataset_Name),len(Data_Type),len(Task),len(Attribute_Type),len( No_of_instances),len(No_of_attribute),len(Year))
```

```python
df=pd.DataFrame({'Dataset Name':Dataset_Name,'Data Type':Data_Type,'Task':Task,'Attribute Type':Attribute_Type,'No of Instances':No_of_instances,'No of Attribute':No_of_attribute,'Year':Year})
df
```
